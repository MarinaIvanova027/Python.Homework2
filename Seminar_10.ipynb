## Задача №67.
Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
- high - длинный(от 42)
- middle - средний(от 35 до 42)
- low - маленький(до 35)

```python
# penguins[penguins['bill_length_mm'] > 42]
penguins.loc[penguins['bill_length_mm'] > 42, 'new_col'] = 'high'
```

```python
penguins
```

```python
penguins.loc[(penguins['bill_length_mm'] <= 42) & (penguins['bill_length_mm'] >= 35), 'new_col'] = 'middle'
penguins
```

```python
penguins.loc[penguins['bill_length_mm'] < 35, 'new_col'] = 'low'
penguins
```

```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
data # без (data) почему-то не показывает название столбца 'whoAmI' ??
```

```python
data.loc[data['whoAmI'] == 'human', 'human'] = '1'
data.loc[data['whoAmI'] != 'human', 'human'] = '0'
```

```python
data.loc[data['whoAmI'] == 'robot', 'robot'] = '1'
data.loc[data['whoAmI'] != 'robot', 'robot'] = '0'
```
